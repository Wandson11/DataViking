```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings

warnings.filterwarnings('ignore')
```

```python
Base = pd.read_excel('/content/Dados_Pib.xlsx')
```

```python
Base.head(10)
```

```python
Base.groupby(by=['Territorialidades']).mean()
```

```python
#sistema de grid


#Cor de fundo
Cor_Fundo = '#f5f5f5'

#Criando o sistema de grid
Grid_Grafico = sns.FacetGrid(Base, col='Territorialidades', hue='Territorialidades', col_wrap=4)
#essa linha de cima, vai criar todo uma estrutura para receber os gráficos individuais de cada estado brasileiro
#assim, o primeiro col recebendo os terrirotorios, fará 28 img (estado + br)
#o col_wrap, indica o numero colunas, que no caso, sao quatro, se nada foir digitado, todos os 28 elementos, se encontrarao em uma unica linha

Grid_Grafico = Grid_Grafico.map( plt.plot, 'Ano', 'PIB per capita')
#plt.plot é um gráfico de linha


#adicionar uma sombra mais ajuste do título
Grid_Grafico = Grid_Grafico.map( plt.fill_between, 'Ano', 'PIB per capita', alpha=0.2).set_titles('{col_name} Territorialidades')
#plt.fill_between, é um gráfico de linha que, pinta a parte debaixo
#alpha indica a opacidade da cor, quanto mais proximo de 1, mais escuro, enquanto o inverso, fica mais claro

#FILTRAR O TITULO, ANTES ESTAVA "ESTADO TERRITORIALIDADE", AGORA FICOU APENAS O ESTADO
#a ideia é tirar o terrotorialidades do titulo de cada gráfico
Grid_Grafico = Grid_Grafico.set_titles('{col_name}')

#ADICIONAR UM TITULO PRINCIPAL PARA ESSE CONJUNTO DE GRÁFICOS
Grid_Grafico = Grid_Grafico.fig.suptitle('PIB por Estado', fontsize=18)

#o titulo geral ficou por cima dos primeiros graficos, USAR O ADJUST
plt.subplots_adjust(top=0.95)
```
