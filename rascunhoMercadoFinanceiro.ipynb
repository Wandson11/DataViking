```python
#bibliotecas para modelagem e matrizes
import pandas as pd
import numpy as np

#bibliotecas para análise gráficas
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

#biblioteca para ignorar aviso
import warnings

#desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
#AQUI É IMPORTANTE SE ATENTAR PARA O TIPO DE DOC QUE FOI DISPONIBILIZADO, AJUSTAR ENTRE CSV, EXCEL, ETC
Base_Dados = pd.read_excel('/content/Vase_004 - Magalu - Sem Resolução.xlsx')
```

```python
#series temporais
#a ideia aqui é transformar a coluna data em linha, para isso, existe o código "set_index('coluna que será convertida')"
Dados = Base_Dados.set_index('Data')
Dados.head()
```

```python
plt.style.use('seaborn-darkgrid') #colocar o tom cinza na parte de tras do gráfico
plt.figure(figsize=(15,7))
plt.title('Análise das Ações da Magalu - Fechamento', fontsize=14, loc='left')
plt.plot(Dados.index, Base_Dados['Fechamento'])
plt.xlabel('Periodo da Cotação')
plt.ylabel('Valor da Ação (R$)');
```

```python
Dados.tail()
```

```python
#ROLLING(x) agrupoa em x em x registros, esse agrupamento pode ser espeficado por um segundo comando
#por exemplo.
Dados['Fechamento'].rolling(5).mean()
```

```python
Media_Movel = Dados['Fechamento'].rolling(5).mean()
Media_Tendencia = Dados['Fechamento'].rolling(30).mean()
```

```python
plt.figure(figsize=(15,6))
plt.style.use('seaborn-dark')
plt.title('Gráfico da variação da Magalu - Fechamento', fontsize=15, loc='left')
plt.plot(Dados.index, Dados['Fechamento'])
plt.plot(Dados.index, Media_Movel)
plt.plot(Dados.index, Media_Tendencia)
plt.xlabel('Período')
plt.ylabel('Valor Ação (R$)');
```

```python
sns.boxplot(data=Dados, x='Fechamento')
```

```python
#boxplot mensal

Dados['Mes'] = Base_Dados['Data'].dt.month #se a data não precisar ser convertido, basta inserir o dt (datetime) e o que se quer, year, day, month
#fiz merda aqui mas deixei
Dados.head()
```

```python
Dados = Dados.drop('Mes', axis=1)
```

```python
Dados.head()
```

```python
#acrescentar na base original

Base_Dados['Mes'] = Base_Dados['Data'].dt.month

Base_Dados.head()
```

```python
plt.figure(figsize=(16,5))
sns.boxplot(data=Base_Dados, x='Mes', y='Fechamento');
```

```python
#com um elemento, não mudou nada a questão de inserir by ou nao
Base_Dados.groupby(['Mes']).describe()['Fechamento']
```

```python
#novo grafico, usando o go
Grafico = go.Figure(
    data=[
        go.Candlestick(  #aqui, o candlestick tem que começar com letra maiuscula,caso contrario, dará erro
            x=Dados.index, 
            open=Dados['Abertura'], 
            high=Dados['Maior'],
            low=Dados['Menor'],
            close=Dados['Fechamento']
        )
    ]
)
Grafico.show()
```
