```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#definir tema.
sns.set_theme(style='darkgrid')
```

```python
#Dados
Banco_Dados = sns.load_dataset('tips')


#Verificar
Banco_Dados.head()
```

```python
#renomear as colunas para portugues
Banco_Dados.rename(columns={'total_bill': 'Total Conta', 
                            'tip': 'Gorjeta',
                            'sex': 'Sexo',
                            'smoker': 'Fumante',
                            'day': 'Dia Semana',
                            'time': 'Periodo', 
                            'size': 'Pessoas na Mesa'}, inplace=True)

#para renomear, basta inserir a variavel que recebeu os dados junto com "rename"
#dentro dos parenteses, chama columns e, inserir chaves, para especficar a key(nome a ser traduzido), com a traducao
```

```python
Banco_Dados.head()
```

# Gráfico Relplot

```python
#no relplot, insere-se o eixo x e y, e ao final, aplica-se de onde esses dados estão retirados por meio de "data"
sns.relplot(x='Total Conta', y = 'Gorjeta', data = Banco_Dados);
```

```python
#relplot igual ao de cima mas com um parametro a mais
#mas agora, fez a separação do sexo masculino para feminino
#hue é a coluna escolhida que separará os dados do replot, deixando-o colorido (categorias diferentes)
sns.relplot(x='Total Conta', y ='Gorjeta', data=Banco_Dados, hue='Sexo')
```

```python
#gráfico de linha com 2 eixos
#ssa mudança é possivel por meio de kind=line
sns.relplot(x='Total Conta', y='Gorjeta', data=Banco_Dados, kind='line')
```

```python
#gráfico de linha com 2 eixos mas separado por fumante e nao fumante
sns.relplot(x='Total Conta', y='Gorjeta', data=Banco_Dados, kind='line', hue='Fumante');
```
