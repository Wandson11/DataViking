# Gráfico de Linhas

```python
np.random.seed(7) #gera um n. aleatório
#pelo o que pesquisei, a linha de código, faz um numero aleatório previsível.

#vetor com n. inteiros aleatórios de 1 a 1500, com 10 amostras

y = np.random.randint(low = 1, high=1500, size = 10)
```

```python
y
```

```python
#fazendo gráfico de linha padrao
plt.plot(y)
```

```python
#insese a primeira linha
plt.plot(y, color = '#749187', marker = 'o', ms=5, mec = 'r', markerfacecolor = 'w', ls = '-.')
#marker é a bolinha que marca o exato local onde o eixo x e y se encontram (para mais opções olhar o doc oficial)
#ms é o tamanho do marcardor
#mec é a cor da borda do marcardor (para mais cores, olhar o doc oficial)
#markerfacecolor é relacionada a cor interna da marcação
#ls é referente a linha em si, onde -. ira fazer uma linha tracejada

#insere a segunda linha
plt.plot(y*2, marker='+', color='k', ms=5)

#rôtulos
plt.xlabel('Eixo X', color='red', size=12)
plt.ylabel('Eixo Y', color='blue', size=12)
plt.title('Título', color='green', size=12, loc='left') #loc é a localização do título

#gridlines
plt.grid(axis = 'y', color='gray', linestyle='--', linewidth = 1, alpha=0.8)
#axis =  é se a grade vem do eixo y, x ou de ambos (both)
#alpha é a opacidade da linha, fica entre 0 e 1
plt.show() #finalizar o gráfico
```

# Subplots

```python
np.random.seed(6)
x = np.arange(1,11)#pelo oque pesquisei, me pareceu um range(1,22,2)
y1 = np.random.randint(1, 400, 10)
y2 = np.random.randint(150, 500, 10)
y3 = np.random.randint(200, 600, 10)

#formatação geral
plt.figure(figsize=(15, 5))
plt.suptitle('Figura', fontsize=15)

# formatação dos subplots
##primeiro gráfico
plt.subplot(1,3,1) # primeira linha, com 3 colunas, o último número, o 1, é a casa em que o gráfico será inserido, linha1/coluna1
plt.plot(x, y1, color='k')
plt.title('Gráfico 1', pad=12)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

##segundo gráfico
plt.subplot(1,3,2)
plt.plot(x, y2, color='r')
plt.title('Gráfico 2', pad=12)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

##terceiro gráfico
plt.subplot(1,3,3)
plt.plot(x, y3, color='g')
plt.title('Gráfico 3', pad=12)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

#teste ara linha de baixo
#plt.subplot(2,3,1)
#plt.plot(x, y2, color='R')
#nao deu certo em inserir uma segunda linha

plt.tight_layout(pad=4) #aumenta o espaçamento entre os gráficos, entre tudo.
plt.show()
```

```python
#fazendo um novo subplot mas de forma que exija menos linhas, porém, corre mais risco de errar
fig, ax = plt.subplots(1, 3, figsize=(15,5))
#o fig, indica ao sistema, que é plt.figure;
#enquanto a variavel ax, recebe as infos dos subplots: de sua quantidade (lin, col) e de suas dimensões
fig.suptitle('Figure Teste') # nomeação do espaço
ax[0].plot(x, y1, color='black')   #ax[0], indica ao programa que está se pedindo para a primeira casa (lin0, col=0)
ax[1].plot(x, y2, color='red') #ax[1], indica que está inserido informações na casa 1 (lin=0, col=1)
ax[2].plot(x, y3, color='green') #casa 2, linha=0 col=2
#lembrete, tanto coluna quanto linha, começam pelo index 0

for c in range(3):
  ax[c].set(title = f'sub-gráfico {c + 1}', xlabel = 'Eixo X', ylabel = 'Eixo Y')
plt.tight_layout(pad=4)
```

```python
#teste meu com o subplotes reduzido
fig, ww = plt.subplots(2, 3, figsize=(20,6))
fig.suptitle('qualquer coisa')
ww[1][1].plot(x, y1, color='orange')
#quanto tem mais de duas linhas, é necessário abrir um segundo colchete, assim, especifica a linha e coluna que o subplot vai entrar
```

```python
import plotly.express as px
```

```python
px.data.gapminder()
```

```python
df = px.data.gapminder().query('country == "Brazil"').set_index('year') #aqui filtrou o dataframe com o query, escolhendo o pais brasil
#antes, existia o index normal, com o set_index, definimos o ano como o índice
#alterando assim, o índice pela coluna year
```

```python
df.head()
```

```python
plt.plot(df.index, df.gdpPercap) #aqui, dava para ter colocado as colunas de df entre colchete, porém, como as mesmas não tinham espaço entre as palavras
#inseri sem colchete mesmo
plt.title('Evolução da renda per capita')
plt.xlabel('Décadas')
plt.ylabel('PIB per capita')
```

```python
df.head()
```

```python
title = ('Expectativa de vida e renda per capita do Brasil')

plt.figure(figsize=(12,5))
plt.scatter(df.lifeExp, df.gdpPercap, marker='D', cmap='viridis')
plt.xlabel('Expectativa de Vida')
plt.ylabel('Renda Per Capita')
plt.title(title, loc='left')
plt.show()
```

Gráfico de Barra

```python
def filtrar_continente(continente):
  df = px.data.gapminder()
  df = df[df['continent'] == continente]
  return df
```

```python
def filtrar_pais(pais, segundavariavel):
  df = px.data.gapminder()
  df = df[df['country'] == pais][segundavariavel]
  return df
```

```python
paises
```

```python
americas = filtrar_continente('Americas')
paises = americas['country'].unique()
plt.figure(figsize=(12, 8))
for pais in paises:
  plt.scatter(filtrar_pais(pais = pais,
                           segundavariavel = 'lifeExp'),
              filtrar_pais(pais = pais,
                           segundavariavel = 'gdpPercap'))
plt.legend(labels = paises, loc='best')
plt.title('Relação entre renda per capita e expectativa de vida no continente Americano', loc='left')
plt.xlabel('Expectativa de vida de cada país')
plt.ylabel('Renda Per Capita');
```
