```python
#Serie Temporal, o ideal é indexar a data
#para isso, existe o set_index
Base.set_index('Data', inplace=True)
```

```python
Base.head()
```

```python
#gráfico

#estilo do gráfico
plt.style.use('seaborn')

plt.figure(figsize=(16,6))

plt.title('Analise das açoes de empresas de energia', fontsize=14, loc='left', fontweight=0)

plt.plot(Base.index, Base['Petrobras'], color='#008c4a', linewidth=4, alpha=0.7)

#plt.text, permite colocar qualquer informaçao dentro do gráfico
plt.text(Base.index[-1], Base['Petrobras'].tail(1), 'Petrobras', color='#008c4a', size='large', horizontalalignment='left')

#plot de todas as colunas
for Coluna in Base.columns[1:]:
  plt.plot(Base.index, Base[Coluna], color='gray', linewidth=2, alpha=0.7)
  #texto das outras cia
  plt.text(Base.index[-1], Base[Coluna].tail(1), Coluna, color='gray')

plt.xlabel('Período')
plt.ylabel('Preço de Fechamento (R$)')
plt.show()
```
